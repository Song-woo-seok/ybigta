```python
import requests
from bs4 import BeautifulSoup
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver'
response= requests.get(url)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
cnt = 0
movies = soup.select('div.tit3')

for movie in movies:
  ans =[]
  ans.append(movie.text.replace('\n',''))

  movie_link = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver' + movie.select_one('a')['href']
  response_movie = requests.get(movie_link) 
  soup_movie = BeautifulSoup(response_movie.text, 'html.parser')

  a = soup_movie.select_one('span.st_on')
  ans.append(a.text)

  info_spec = soup_movie.select_one('dl.info_spec')
  movie_details = info_spec.select('dd > p')
 
  for movie_detail in movie_details:
    if movie_detail.select('span'):
      for i in movie_detail.select('span'):
        ans.append(i.text.replace('\n','').replace('\t','').replace('\r',''))
    else:
      ans.append(movie_detail.text.replace('\n','').replace('\t','').replace('\r',''))

  movie_people = info_spec.select('p.count')
  for i in movie_people:
    ans.append(i.text.replace('\n','').replace('\t','').replace('\r',''))
  print(ans)

  cnt += 1
  if cnt == 10:
    break
```
